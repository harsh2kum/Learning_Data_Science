```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv(r"D:\Learning Data Science\Dataset\imdb-top-1000.csv")
movies
```

```python
movies.head()
```

```python
#Groupby is apply on catogorical data
genres = movies.groupby('Genre')
genres
```

```python
#Applying builtin aggregation function on groupby objects 
genres.sum()
genres.min()
genres.max()
```

```python
# find the top 3 genres by total earning
movies.groupby('Genre').sum()['Gross'].sort_values(ascending=False).head(3)
```

```python
movies.groupby('Genre')['Gross'].sum().sort_values(ascending=False).head(3)
```

```python
# find the genre with highest avg IMDB rating
movies.groupby('Genre')['IMDB_Rating'].mean().sort_values(ascending=False).head(1)
```

```python
# find director with most popularity
movies.groupby('Director')['No_of_Votes'].sum().sort_values(ascending=False).head(1)
```

```python
# find the highest rated movie of each genre
movies.groupby('Genre')['IMDB_Rating'].max()
```

```python
# find number of movies done by each actor
movies['Star1'].value_counts()
#or
movies.groupby('Star1')['Series_Title'].count().sort_values(ascending=False)
```

```python
# GroupBy Attributes and Methods
# find total number of groups -> len
# find items in each group -> size
# first()/last() -> nth item
# get_group -> vs filtering
# groups
# describe
# sample
# nunique
```

```python
# find total number of groups -> len
len(movies.groupby('Genre'))
#or 
movies['Genre'].nunique()
```

```python
# find items in each group -> size
movies.groupby('Genre').size()
```

```python
# first()/last() -> nth item
genres = movies.groupby('Genre')
genres.first()
genres.last()
genres.nth(6)
```

```python
# get_group -> vs filtering
#if i want to check particular genre we use get_group it is fast 
genres.get_group('Horror')
genres.get_group('Fantasy')

#or not use get_group, 2nd method 
movies[movies['Genre'] == 'Fantasy']

```

```python
#ALL GROUPS
genres.groups
```

```python
genres.describe()
genres.sample()
genres.sample(3,replace = True) #3 random movies of each genre 
genres.nunique()
```

```python
# Aggregate method
movies.head(1)
genres.agg(
    {
        'Runtime':"mean",
        'IMDB_Rating': 'mean',
        'No_of_Votes': 'sum',
        'Gross':'sum',
        'Metascore':'min'
    }
)
```

```python
#pagging list 
genres.agg(['min','max','sum'])
```

```python
genres.agg(
    {
        'Runtime':["mean",'min'],
        'IMDB_Rating': ['mean','sum'],
        'No_of_Votes': 'sum',
        'Gross':'sum',
        'Metascore':'min'
    }
)
```

```python
# looping on groups
df = pd.DataFrame(columns=movies.columns)

for group, data in genres:
    #we use concat in pandas instead of apppend 
    df = pd.concat([df, data[data['IMDB_Rating'] == data['IMDB_Rating'].max()]], ignore_index=True)

df

# - Replaced df.append(...) with pd.concat([...], ignore_index=True) — this avoids repeated DataFrame creation and is more efficient
```

```python
#find the highest rated movies of each genre
for group,data in genres:
    print(data[data["IMDB_Rating"] == data["IMDB_Rating"].max()])
```
