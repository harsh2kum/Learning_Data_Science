```python
#drop Duplicates
temp = pd.Series([1,1,2,2,3,3,4,4])
temp.drop_duplicates() #drop 1st occurence #by Default 
temp.drop_duplicates(keep='last') ##drop last occurence

temp.duplicated() #to check duplicate value exist or not
temp.duplicated().sum() #tells total duplicate value
```

```python
vk.duplicated().sum()
movies.drop_duplicates()
```

```python
temp = pd.Series([1,2,3,np.nan,5,6,np.nan,8,np.nan,10])
temp
temp.size #total values = 10
temp.count() #give nit null values which is 7
```

```python
#isnull - to check missing or null values
vk.isnull()
vk.isnull().sum()
# temp.isnull().sum()
```

```python
#dropna- remove missing values or null values
temp.dropna()
```

```python
#fillna - use to fill null or missing values 
temp.fillna(0) #fill all the missing values with 0
temp.fillna(temp.mean()) ##fill all the missing values with mean value
```
