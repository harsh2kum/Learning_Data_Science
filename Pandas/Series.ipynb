### What is Pandas
Pandas is a fast, powerful, flexible and easy to use open source data analysis and manipulation tool, built on top of the Python programming language.

### Pandas Series
A Pandas Series is like a column in a table. It is a 1-D array holding data of any type.

```python
import numpy as np
import pandas as pd
```

#### Series form list

```python
#string
country  = ['india','usa','nepal','srilanka']
pd.Series(country)
```

```python
#integer 
runs = [13,15,23,35,44,55,100]
pd.Series(runs)
```

```python
#customer index
marks = [66,75,85,100]
subjects = ['maths','science','hindi','english']
pd.Series(marks,subjects)
#or
pd.Series(marks,index= subjects)
```

```python
#setting a name
marks = pd.Series(marks,index=subjects,name='college marks')
marks
```

```python
#series from dict
marks = {
    'marks':60,
    'english':69,
    'science':79,
    'hindi': 50
}
m = pd.Series(marks,name="Harsh")
m
```

```python
#Series Attributes 
#size
m.size
#dtype
m.dtype
#name
m.name
#is_unique
m.is_unique #output-True
# pd.Series([1,1,2,3,4,5]).is_unique #output - False
#index
m.index
#values
m.values
```

```python
# sort_values 
"""sort_values() method in pandas needs to know which column to sort by. 
That’s what the by parameter is for.
#syntax-- vk.sort_values(by='column_name')
"""
# vk.columns
vk.sort_values(by = 'match_no')
vk.sort_values(by = 'runs',ascending=False) # sort in descending order
vk.sort_values(by=['match_no', 'runs'], ascending=[False, True]) #sort by multiple columns
vk.sort_values(by=['runs'],ascending=False).head(1).values[0] #method chaining 

```

### What is method chaining -- 
Method chaining in pandas is a powerful and elegant way to write concise, readable code by linking multiple operations together in a single line or block. It’s like building a pipeline where each step transforms the data and passes it to the next.


```python
#for example
#Instead of writing:
# df = df.dropna()
# df = df.sort_values(by='views')
# df = df.reset_index(drop=True)
# # You can chain them
# df = (df.dropna().sort_values(by='views').reset_index(drop=True))

```
