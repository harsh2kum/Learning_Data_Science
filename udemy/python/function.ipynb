#### *args and **kwargs are special keywords that are used to pass the variable length of arguments to a function 
#### *args = allows us to pass a variable number of non-keyword arguments to a function 

```python
# *args
def mult(*args): #we can use any name in place of *args
    prod = 1
    for i in args:
        prod = prod * i
        
    return prod 

res = mult(1,2,3,4,5,9,8) #any number of arguments can be given 
print(res) 


def mult(*harsh):
    prod = 1
    for i in harsh:
        prod = prod * i
        
    return prod 

res = mult(1,2,3,4,5,9,8) #any number of arguments can be given 
print(res)
```

#### **kwargs - allows us to pass any number of keyword arguments 
#### keyword arguments mean that they contain a key-value pair , like a Python dictionary

```python
def display(**kwargs): #we can use any name in place of **kwargs
    for (key,val) in kwargs.items():
        print(f'{key} -> {val}')
    
#call 
display(India ="delhi",Srilanka='colombo',Russia = "Mossco")

# **kwargs convert input into dictionary
 
```

### Lambda Function 
A lambda function is a small anonymous function , A lambda function can take any number of arguments, but can only have one expression.



### Diff b/w lambda vs Normal Function 
- No name 
- lambda has no return value (infact,returns a function)
- lambda is written in 1 line
- not resuable

```python
a = lambda x: x**2
a(4)
b = lambda x,y:x+y
b(3,5)
```

```python
#check if s string has 'a'
c = lambda s : 'y' in s
c("python")

#odd or even 
a = lambda x:'even' if x%2 == 0 else 'odd'
a(8)
```

Higher Order Function(HOF) - using lambda

```python
L = [1,2,3,4,5]
def square(x):
    return x**2

def trasnsform(f,L): #HOF
    output = []
    for i in L:
        output.append(f(i))
    print(output)

#call the function
trasnsform(square, L)  
trasnsform(lambda x:x**3,L)
```

### Enumerate Function in Python

##### The enumerate() function in Python is a built-in utility that adds a counter to an iterable and returns it as an enumerate object. It's especially useful when you need both the index and the value while looping through a list, tuple, or string.

#### Syntax - enumerate(iterable, start=0)
- iterable: Any iterable object (like a list, tuple, string).
- start: The starting index (default is 0).\
### why use enumerate
- Cleaner than manually managing an index with range(len(...))
- More Pythonic and readable
- Works with any iterable, not just lists



```python
fruits = ['apple', 'banana', 'cherry']
for index, fruit in enumerate(fruits):
    print(index, fruit)
    
#Custom Start Index
for index, fruit in enumerate(fruits, start=1):
    print(index, fruit)
    
    
#we can convert the result to a list or tuple:
list(enumerate(fruits))

```
