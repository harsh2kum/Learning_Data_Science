 File operation - Read and Write files


```python
# Read a Whole file 
with open('example.txt','r') as file:
    content = file.read()
    print(content)
```

```python
# Read a file line by line 
with open('example.txt','r') as file:
    for line in file:
        print(line.strip()) #.strip() remove the newline character
```

```python
# Write a file and (overwriting)

with open('example.txt','w') as file:
    file.write("Hello how are you?\n")
    file.write("This is a new line\n")
    file.write("my name is harsh kumar gupta \n")
    file.write("welcome to hyderabad. \n")
```

```python
#write a file(without overwritten) (Appending )
with open('example.txt','a') as file:
    file.write("Append operation taking place!\n")
```

```python
## writing a list of lines to a file 
lines = ['First line \n','second line \n','Third line\n']
with open('example.txt','a') as file:
    file.writelines(lines)
```

```python
# Binary Files 

#write to a binary file 
data = b'\x00\x01\x02\x03\x04'
with open('example.bin','wb') as file:
    file.write(data)

```

```python
# Reading a binary file 
with open('example.bin','rb') as file:
    content = file.read()
    print(content)
```

```python
# Read the content from a source text file and write to a destination text file

#copying a text file 
with open('example.txt','r') as source_file:
    content = source_file.read()
    
with open('destination.txt','w') as destination_file:
    destination_file.write(content)
```

```python
#Read text file and count the number of lines,words and characters

#counting lines , words and characters in a text file

def count_text_file(file_path):
    with open(file_path,'r') as file:
        lines = file.readlines()
        line_count = len(lines)
        word_count = sum(len(line.split()) for line in lines)
        char_count = sum(len(line) for line in lines)
    return line_count,word_count,char_count

file_path = 'example.txt'
lines,words,characters = count_text_file(file_path)
print(f'Lines: {lines}, Words: {words},characters: {characters}')
```

The w+ mode in python is used to open a file for both reading and writing. if the file does not exist, it will be created. if the file exists,its content is truncated (i.e., the file is overwritten)

```python
# writing and then reading a file 

with open('example.txt','w+') as file:
    file.write("Hello world \n")
    file.write("This is a new line \n")
    
  #move the file cursor to the beginning 
    file.seek(0)
  
  # read the content of file 
    content = file.read()
    print(content)
```

```python
#Working with File Paths
#Creating a New Directory 

import os
new_directory = "package"
os.mkdir(new_directory)
print(f'Directory {new_directory} is created.')
```

```python
# Joining Paths

directory_name = "file.txt"
file_name = "file.txt"
full_path = os.path.join(directory_name, file_name)
print(full_path)
```

```python
# To get the full path including the current working directory, use the following:

import os
full_path = os.path.join(os.getcwd(), directory_name, file_name)
print(full_path)

#This prints the complete path. If only the relative path is needed, use just the directory and file name variables.
```

```python
# Checking If a Path Exists

example_file = "example1.txt"
if os.path.exists(example_file):
    print(f'The path {example_file} exists.')
else:
    print(f"The path {example_file} does not exist.")
    
    #If the file does not exist, you can create it using the open command or any other file creation method. This kind of condition is important when making assumptions about specific paths in your code.
    with open(example_file, "w") as file:
        file.write("This is a newly created file.\n")
    print(f"{example_file} has been created.")
    
```

```python
#Checking if a path is a file or Directory
#There is another way to check whether a path is a file or a directory.
path = "example.txt"
if os.path.isfile(path):
    print(f"The path {path} is a file.")
elif os.path.isdir(path):
    print(f"The path {path} is a directory.")
else:
    print(f"The path {path} is neither a file nor a directory.")
```

```python
# Relative and Absolute Paths
# A relative path refers to a file inside a specific folder, such as file.txt inside a folder. An absolute path is the complete path from the root directory.
relative_path = "example.txt"
absolute_path = os.path.abspath(relative_path)
print(absolute_path)
```
